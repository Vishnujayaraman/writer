## Import all the required python libraries 

```python
import warnings
warnings.simplefilter("ignore")
```

```python
import numpy as np
import pandas as pd
```

```python
from sklearn.model_selection import train_test_split
```

## Import and View Dataset

```python
dataset = pd.read_csv("Admission_Predict_Ver1.1.csv")
```

```python
dataset
```

## Remove Unnnecessary columns

```python
datset = dataset.drop(['Serial No.', 'TOEFL Score', 'University Rating', 'SOP', 'LOR ', 'CGPA', 'Research'], axis=1)
```

```python
dataset
```

## Reshape the dataset

```python
x = dataset.iloc[:,0].values.reshape(-1,1) 
y = dataset.iloc[:,-1].values.reshape(-1,1) 
```

## Divide dataset into training set and testing set

```python
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

## Import linear regression class 

```python
from sklearn.linear_model import LinearRegression 
```

## Create an object of the linear regression class

```python
lm = LinearRegression()
```

## Fitting the data 

```python
lm.fit(x_train,y_train) 
```

## Predicting the output 

```python
y_pred = lm.predict(x_test) 
```
