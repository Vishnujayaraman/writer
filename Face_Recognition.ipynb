## Convert the dataset into Data frame

```python
Akshay_kumar_data = pd.DataFrame(Akshay_kumar_gray) 
Alia_bhatt_data = pd.DataFrame(Alia_bhatt_gray)
Anushak_sharma_data = pd.DataFrame(Anushak_sharma_gray)
Priyanka_chopra_data = pd.DataFrame(Priyanka_chopra_gray)
```

## Add label to the dataset

```python
Akshay_kumar_data["label"] = "Akshay_Kumar"
Alia_bhatt_data["label"] = "Alia_Bhatt"
Anushak_sharma_data["label"] = "Anushak_Sharma"
Priyanka_chopra_data["label"] = "Priyanka_Chopra"
```

## Merge Data frames

## Shuffle the dataset

```python
from sklearn.utils import shuffle
```

```python
kollywood_indexed = shuffle(actor).reset_index()
```

```python
kollywood_indexed
```

## Remove Index

```python
kollywood_actors=kollywood_indexed.drop(['index'],axis=1)
```

```python
kollywood_actors
```

## Identify dependent and independent data

```python
x=kollywood_actors.values[:,:-1]
y=kollywood_actors.values[:,-1]
```

## Divide the dataset into Training set and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

## Import PCA model

```python
from sklearn import decomposition
```

## Fit the PCA model with independent data

```python
pca = decomposition.PCA(n_components =70, whiten=True)
```

```python
x_train.shape
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=0.95)  # Keep enough components to explain 95% of variance
pca.fit(x_train)
```

## Extract Eigen Components

```python
X_train_pca = pca.transform(x_train)
X_test_pca = pca.transform(x_test)
```

```python
X_train_pca.shape
```

## Fit Data into support vector machines model

```python
from sklearn import svm
```

```python
clf = svm.SVC(C=2., gamma=0.006)
clf.fit(X_train_pca, y_train)
```

## Perform prediction using Test set

```python
y_pred = clf.predict(X_test_pca)
```
