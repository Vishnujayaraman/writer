## Convert Images to Gray Scale

```python
Hrithik_Roshan_gray = [None]*limit
Jessica_Alba_gray = [None]*limit
Priyanka_Chopra_gray = [None]*limit
Vijay_Deverakonda_gray = [None]*limit
```

```python
j = 0
while j < limit:
    Hrithik_Roshan_gray[j] = rgb2gray(Hrithik_Roshan_images[j])
    Jessica_Alba_gray[j] = rgb2gray(Jessica_Alba_images[j])
    Priyanka_Chopra_gray[j] = rgb2gray(Priyanka_Chopra_images[j])
    Vijay_Deverakonda_gray[j] = rgb2gray(Vijay_Deverakonda_images[j])
    j+=1
```

## Resize the image

## Matrix to Vector Conversion

## Flatten the images

```python
len_Hrithik_Roshan = len(Hrithik_Roshan_gray)
len_Jessica_Alba = len(Jessica_Alba_gray)
len_Priyanka_Chopra = len(Priyanka_Chopra_gray)
len_Vijay_Deverakonda = len(Vijay_Deverakonda_gray)
```

```python
image_size = (160, 160)
```

```python
flatten_size=image_size[0]*image_size[1]
```

```python
flatten_size
```

```python
for i in range(limit):
    Hrithik_Roshan_gray[i] = np.ndarray.flatten(Hrithik_Roshan_gray[i]).reshape(flatten_size)
    Jessica_Alba_gray[i] = np.ndarray.flatten(Jessica_Alba_gray[i]).reshape(flatten_size)
    Priyanka_Chopra_gray[i] = np.ndarray.flatten(Priyanka_Chopra_gray[i]).reshape(flatten_size)
    Vijay_Deverakonda_gray[i] = np.ndarray.flatten(Vijay_Deverakonda_gray[i]).reshape(flatten_size)
```

## Creating Data set

```python
Hrithik_Roshan_gray = np.dstack(Hrithik_Roshan_gray)
Hrithik_Roshan_gray = np.rollaxis(Hrithik_Roshan_gray, axis=2, start=0)
Hrithik_Roshan_gray = Hrithik_Roshan_gray.reshape(limit, flatten_size)
Hrithik_Roshan_data = pd.DataFrame(Hrithik_Roshan_gray)
Hrithik_Roshan_data["label"] = "Hrithik_Roshan"
```

```python
Jessica_Alba_gray = np.dstack(Jessica_Alba_gray)
Jessica_Alba_gray = np.rollaxis(Jessica_Alba_gray, axis=2, start=0)
Jessica_Alba_gray = Jessica_Alba_gray.reshape(limit, flatten_size)
Jessica_Alba_data = pd.DataFrame(Jessica_Alba_gray)
Jessica_Alba_data["label"] = "Jessica_Alba"
```

```python
Priyanka_Chopra_gray = np.dstack(Priyanka_Chopra_gray)
Priyanka_Chopra_gray = np.rollaxis(Priyanka_Chopra_gray, axis=2, start=0)
Priyanka_Chopra_gray = Priyanka_Chopra_gray.reshape(limit, flatten_size)
Priyanka_Chopra_data = pd.DataFrame(Priyanka_Chopra_gray)
Priyanka_Chopra_data["label"] = "Priyanka_Chopra"
```

```python
Vijay_Deverakonda_gray = np.dstack(Vijay_Deverakonda_gray)
Vijay_Deverakonda_gray = np.rollaxis(Vijay_Deverakonda_gray, axis=2, start=0)
Vijay_Deverakonda_gray = Vijay_Deverakonda_gray.reshape(limit, flatten_size)
Vijay_Deverakonda_data = pd.DataFrame(Vijay_Deverakonda_gray)
Vijay_Deverakonda_data["label"] = "Vijay_Deverakonda"
```

## Creating Actor data set

```python
actor = pd.concat([Hrithik_Roshan_data, Jessica_Alba_data, Priyanka_Chopra_data, Vijay_Deverakonda_data])
```

```python
actor
```

## Shuffing within Final dataframe

```python
from sklearn.utils import shuffle
```

```python
kollywood_indexed = shuffle(actor).reset_index()
kollywood_indexed
```

## Removing Index from Final Dataset

```python
kollywood_actors=kollywood_indexed.drop(['index'],axis=1)
kollywood_actors
```

## Assigning Dependent and Independent Variables

```python
x=kollywood_actors.values[:,:-1]
y=kollywood_actors.values[:,-1]
```

## Splitting Data

## Divide the dataset into Training set and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.5,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
```

## Import PCA model
## Principle Componant Analysis(PCA) Algorithm

```python
from sklearn import decomposition
pca = decomposition.PCA(n_components=10, whiten=True, random_state=1)
```

```python
pca.fit(x_train)
```

```python
x_train_pca = pca.transform(x_train)
x_test_pca = pca.transform(x_test)
```

```python
x_train_pca.shape
```

```python
x_test_pca.shape
```

## Viewing the principle components or eigen faces

```python
eigen = (np.reshape(x[10],(160,160)).astype(np.float64))
```

```python
eigen.shape
```

## Plotting images one by one as subplots

```python
fig = plt.figure(figsize=(30, 30))
for i in range(10):
    ax = fig.add_subplot(2, 5, i + 1, xticks=[], yticks=[])
    ax.imshow(pca.components_[i].reshape(eigen.shape), cmap=plt.cm.bone)
    ax.set_title(f'Principal Component {i + 1}', fontsize=20)
plt.show()
```

## Fit Data into support vector machines model

## Support Vector Machines Algorithm

```python
from sklearn import svm
```

### Classification

```python
#Classification
clf = svm.SVC(C=2, gamma=0.006, kernel='rbf')
clf.fit(x_train_pca, y_train)
```

## Image Prediction

```python
y_pred = clf.predict(x_test_pca)
```

## Display Predicted out

```python
for i in (np.random.randint(0,10,10)):
    predicted_images = (np.reshape(x_test[i], (160,160)).astype(np.float64))
    plt.title('predicted label: {0}'. format(y_pred[i]))
    plt.imshow(predicted_images, interpolation='nearest', cmap='gray')
    plt.show()
```

## Prediction Accuracy

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,y_pred)
accuracy
```

```python
from sklearn.metrics import confusion_matrix, classification_report
confusion_matrix(y_test,y_pred)
```

```python
print(classification_report(y_test, y_pred))
```
