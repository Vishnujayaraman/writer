## Handling Categorical Data

```python
# Importing Libraries
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
# Creating data
salary_class = pd.DataFrame({'Salary':[5000,84000,22000,8000,75000], 'Class':['Low','High','Medium','Low','High']})
```

```python
salary_class
```

```python
# 5.3.3. Applying label encoder 
lab_encode = LabelEncoder()
```

```python
# 5.3.4. Fitting the values and transforming
salary_class['Class'] = lab_encode.fit_transform(salary_class['Class'])
```

```python
# Label encoded output
salary_class
```

## Feature Scaling

```python
# Creating data frame for minmax scaler

mms = pd.DataFrame({'x1':np.random.randint(0,10,size=100),
                    'x2':np.random.randint(-10,10,size=100)*40,
                    'x3':np.random.randint(50,100,size=100)})
```

```python
mms
```

```python
# Importing Minmaxscaler
from sklearn.preprocessing import MinMaxScaler
```

```python
# Create an object with min max scaler
minmax = MinMaxScaler()
```

```python
# Fitting & viewing data frome in the object
mms = minmax.fit_transform(mms)
```

```python
mms
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.kdeplot(mms)
plt.show()
```
