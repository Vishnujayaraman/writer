## Creating data

```python
Days = ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
Temperature = [33,34,37,36,39,31,32]
```

```python
Movies = ['JL','Avengers','Batman','Ironman','Mask']
Percentage = [23,30,21,19,7]
```

## Data Visualization using Matplotlib

```python
# Scatter Plot
plt.scatter(Days, Temperature)

# Adding labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.title('Temperature throughout the Week')

# show plot
plt.show()
```

```python
# Line Plot
plt.plot(Days, Temperature, linestyle='solid')

# Adding labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.title('Temperature throughout the Week')

# show plot
plt.show()
```

```python
# Line Plot with Dotted line
plt.plot(Days, Temperature, linestyle='dotted')

# Adding labels and title
plt.xlabel('Day of the Week')
plt.ylabel('Temperature (°C)')
plt.title('Temperature throughout the Week')

# show plot
plt.show()
```

```python
# Pie Chart
plt.pie(Percentage, labels=Movies)
```

```python
# Bar Plot 
plt.bar(Movies, Percentage)
```

## Importing Seaborn Library

### Importing Seaborn Library and other required libraries

```python
# Check for existing default datasets in seaborn
sns.get_dataset_names()
```

```python
# Loading tips dataset
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
# Relational Plot
sns.relplot(x='total_bill',y='tip',data=tips)
```

```python
# Categorical Plot
sns.catplot(x='day',y='total_bill',data=tips)
```

```python
# Bar Plot
sns.catplot(x='day',y='total_bill',kind='box',data=tips)
```
